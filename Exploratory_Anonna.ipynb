```python
#importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
#loading data into dataframe
df=pd.read_excel("/content/project.xls.xlsx")
#checking data types
df.dtypes
```

```python
#Dropping irrelevent columns
df=df.drop(["Favorite Color"],axis=1)
print (df)
```

```python
#renaming the columns
df=df.rename(columns={"Favorite Food":"Food"})
print(df)
```

```python
#Dropping the duplicate rows
total_rows =len(df)#before removing duplicate rows
print("Total number of rows before removing duplicate rows",total_rows)
df=df.drop_duplicates()
total_rows=len(df)#after removing duplicates
print("Total number of rows after removing duplicate rows",total_rows)
```

```python
#Finding the null values
print(df.isnull().sum())
```

```python
#Dropping the missing vlues
df=df.dropna()
print(len(df))
```

```python
#Measuring the value of IQR
numerical_df=df.select_dtypes(include=np.number)
Q1=numerical_df.quantile(0.25)
Q3=numerical_df.quantile(0.75)
IQR=Q3-Q1
print(IQR)
```

```python
df=numerical_df[~((numerical_df<(Q1-1.5*IQR)) | (numerical_df>(Q3+1.5*IQR))).any(axis=1)]
df.shape
```

```python
#plotting a histogram
df.Math.plot(kind="bar",figsize=(10,5))
plt.title("Students marks in Math")
plt.xlabel("Name")
plt.ylabel("Math")
```
